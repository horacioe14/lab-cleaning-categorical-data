```python
import pandas as pd
import numpy as np
import datetime
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head()
```

```python
data.columns = data.columns.str.lower() # lower case
data.columns = map(lambda x : x.replace('-', '_').replace(' ', '_'), data.columns) 
data.head()
```

```python
data.dtypes
```

```python
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'], errors = 'coerce' )
data.head()
```

```python
data.isna().sum()
```

```python
class preprocessing():
    
    def funk(self, data):
        self.categoricals_df = data.select_dtypes('object')
        self.discrete_df = data.select_dtypes('int64')
        self.continuous_df = data.select_dtypes('float64')
    
    def nulls(self, data):
        is_na = categoricals_df.isna()
        return is_na
```

```python
# Load the continuous and discrete variables into continuous_df and discrete_df variables.
prep = preprocessing()
prep.funk(data)
prep.discrete_df.columns 
#prep.continuous_df


```

```python
numerical = pd.concat([prep.discrete_df,prep.continuous_df], axis = 1)
```

```python
# Plot a correlation matrix, what can you see?
mask = np.zeros_like(numerical.corr())
mask[np.triu_indices_from(mask)] = True # optional, to hide repeat half of the matrix
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(numerical.corr(), mask=mask, annot=True)
plt.show()
```

```python
# Create a function to plot every discrete variables. Do the same with continuous variables (be careful, you may change the plot type to another one better suited for continuous data).
```

```python
def plot_discrete(data):
    for col in data:
        sns.distplot(data[col])
        plt.show()

plot_discrete(prep.discrete_df)
```

```python
def plot_continuous(data):
    for col in data:
        sns.distplot(data[col])
        plt.show()

plot_continuous(prep.continuous_df)
```
